# Creation of Recency, Frequency of orders and Total Amount in preparation for RFM segmentation.

RFM segmentation takes into account Recency (date of last order), Frequency of orders and Amount (of last order or over a given period) to establish homogeneous customer segments.

```python
# Selection of interesting columns

Selected_Variables = Combined_Dataset[['customer_unique_id', 'price', 'payment_value', 'order_purchase_timestamp', 'payment_sequential', 'payment_type', 'payment_installments', 'review_score', 'customer_zip_code_prefix', 'product_category_name_english']]
```

```python
Selected_Variables.info()
```

```python
# Dropping 3 NaN's

Selected_Variables = Selected_Variables.dropna(subset = ['payment_value'])
```

```python
# Filling review score with 3, ie average

Selected_Variables['review_score'] = Selected_Variables['review_score'].fillna(value=3)
```

```python
# Filling unknown categories with unknown

Selected_Variables['product_category_name_english'] = Selected_Variables['product_category_name_english'].fillna('unknown')
```

```python
# Searching for the total number of customer_unique_id

counts = Selected_Variables['customer_unique_id'].value_counts()
```

```python
# printing first 100

counts.head(100)
```

```python
# Understanding that there can be four data rows for one customer purchase

Selected_Variables.loc[Selected_Variables['customer_unique_id'] == '1da09dd64e235e7c2f29a4faff33535c']
```

```python
# Dropping data to leave one row per customer purchase

Selected_Variables.drop_duplicates(subset=['customer_unique_id', 'order_purchase_timestamp'], inplace=True, ignore_index=True)
```

```python
# Confirming dropping data has worked as expected

Selected_Variables.loc[Selected_Variables['customer_unique_id'] == '1da09dd64e235e7c2f29a4faff33535c']
```

```python
# Calculation of the number of times one customer has made a purchase

Frequency = Selected_Variables['customer_unique_id'].value_counts().rename_axis('customer_unique_id').reset_index(name='frequency')
```

```python
Frequency.head()
```

```python
Frequency.info()
```

```python
# Adding frequency to the dataframe

Selected_Variables = pd.merge(Selected_Variables, Frequency, how="left", on="customer_unique_id")
```

```python
# Checking for no NaN values

Selected_Variables.info()
```

```python
# Calculation of total spend per customer

Sum_of_Prices = Selected_Variables.groupby('customer_unique_id').agg({'price':'sum'}).reset_index()
```

```python
# Column rename

Sum_of_Prices = Sum_of_Prices.rename(columns={'price': 'Sum_of_Prices'})
```

```python
Sum_of_Prices.head()
```

```python
# Adding total sum to the dataframe

Selected_Variables = pd.merge(Sum_of_Prices, Selected_Variables, how="left", on="customer_unique_id")
```

```python
Selected_Variables.head()
```

```python
# Converting order_purchase_timestamp to datetime

Selected_Variables['order_purchase_timestamp'] = pd.to_datetime(Selected_Variables['order_purchase_timestamp'])
```

```python
Selected_Variables.sort_values("order_purchase_timestamp", ascending = False)
```

```python
# Selection of best customer to double check sum of prices and frequency

Selected_Variables.loc[Selected_Variables['customer_unique_id'] == '8d50f5eadf50201ccdcedfb9e2ac8455']
```

```python
# Calculation of average price paid per customer

Selected_Variables['Average_Price'] = Selected_Variables['Sum_of_Prices']/Selected_Variables['frequency']
```

```python
# Sorting via order_purchase_timestamp for a specfic customer

Selected_Variables.loc[Selected_Variables['customer_unique_id'] == '8d50f5eadf50201ccdcedfb9e2ac8455'].sort_values('order_purchase_timestamp', ascending=False).head()
```

```python
# Collection of the data with the most recent date a customer has made a purchase 

Last_Order_Date = Selected_Variables.groupby('customer_unique_id').agg({'order_purchase_timestamp':'max'}).reset_index()
```

```python
Last_Order_Date.head()
```

```python
Last_Order_Date.info()
```

```python
# Creating a new column and assigning it the date of 2/11/2018 

Last_Order_Date['last_date'] = date.fromordinal(737000)
```

```python
# Converting to datetime

Last_Order_Date['last_date'] = pd.to_datetime(Last_Order_Date['last_date'])
```

```python
# Total time between the last order time and the 2/11/2018

Last_Order_Date['Difference'] = Last_Order_Date['last_date'] - Last_Order_Date['order_purchase_timestamp']
```

```python
# Conversion of difference into seconds

Last_Order_Date['Seconds_Since_Last_Order'] = Last_Order_Date['Difference'].dt.total_seconds()
```

```python
# To avoid getting two customer_unique_id columns

Selected_Variables = Selected_Variables.drop(columns=['customer_unique_id'])
```

```python
Order_Dataset = pd.merge(Last_Order_Date, Selected_Variables, how="left", on="order_purchase_timestamp")
```

```python
Order_Dataset = Order_Dataset.rename(columns={'order_purchase_timestamp': 'Date_of_last_purchase'})
```
